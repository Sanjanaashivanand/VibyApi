```python
ExcitedPlaylist = pd.read_csv('./Data/Training/Excited.csv')
DelightedPlaylist = pd.read_csv('./Data/Training/Delighted.csv')
HappyPlaylist = pd.read_csv('./Data/Training/Happy.csv')
ContentPlaylist = pd.read_csv('./Data/Training/Content.csv')
CalmPlaylist = pd.read_csv('./Data/Training/Calm.csv')
SleepPlaylist = pd.read_csv('./Data/Training/Sleepy.csv')
DepressedPlaylist = pd.read_csv('./Data/Training/Depressed.csv')
SadPlaylist = pd.read_csv('./Data/Training/Sad.csv')
AngryPlaylist = pd.read_csv('./Data/Training/Angry.csv')

TrainingDataset = pd.concat([ExcitedPlaylist,DelightedPlaylist,HappyPlaylist,
                             ContentPlaylist,CalmPlaylist,SleepPlaylist,
                            DepressedPlaylist,SadPlaylist,AngryPlaylist], ignore_index=True)
TrainingDataset
```

```python
TrainingFeatures = scf.getFeatures(TrainingDataset)
TrainingFeatures
```

```python
TrainingDF = TrainingDataset.merge(TrainingFeatures, on="id", how='left')
```
